```python
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
data_path = '../../data/metadata/'
```

Kudos distribution in fandoms

```python
for f in os.listdir(data_path)[0:2]:
    plt.figure()
    df = pd.read_csv(os.path.join(data_path, f), sep = '\t')
    df = df.drop(['ChapterIndex', 'URL','CompleteDate','UpdateDate'], axis=1)
    df = df.drop_duplicates()
    kudos = [i for i in df.Kudos.tolist() if not np.isnan(i)]
    plt.hist(sorted(kudos))
    plt.title(f)
```

Few authors get all kudos

```python
for f in os.listdir(data_path)[0:2]:
    plt.figure()
    df = pd.read_csv(os.path.join(data_path, f), sep = '\t')
    df = df.drop(['ChapterIndex', 'URL','CompleteDate','UpdateDate'], axis=1)
    df = df.drop_duplicates()
    
    df = df.groupby(by=['Author'])['Hits'].sum().reset_index()
    hits = [i for i in df.Hits.tolist() if not np.isnan(i)]
    plt.hist(sorted(hits))
    plt.xlabel('Hits')
    plt.ylabel('Authors')
    plt.title(f.split('.')[0].replace("_metadata", ''))
```

```python
df.fillna(0).sort('Kudos').tail()
```

Cumulative

```python
for f in os.listdir(data_path)[0:2]:
    plt.figure()
    df = pd.read_csv(os.path.join(data_path, f), sep = '\t')
    print(df.columns.values)
    df = df.drop(['ChapterIndex', 'URL','CompleteDate','UpdateDate'], axis=1)
    df = df.drop_duplicates()
    
    df = df.groupby(by=['Author'])['Kudos'].sum().reset_index()

    kudos = [i for i in df.Kudos.tolist() if not np.isnan(i)]
    X = sorted(kudos)
    N = len(X)
    Y = np.linspace( 1/N, 1, num=N)
    plt.plot(X,Y)
    plt.xlabel('Kudos')
    plt.ylabel('Percentage')
    plt.title(f.split('.')[0].replace("_metadata", ''))

```

Same with hits

```python
for f in os.listdir(data_path)[0:2]:
    plt.figure()
    df = pd.read_csv(os.path.join(data_path, f), sep = '\t')
    df = df.drop(['ChapterIndex', 'URL','CompleteDate','UpdateDate'], axis=1)
    df = df.drop_duplicates()
    
    df = df.groupby(by=['Author'])['Hits'].sum().reset_index()

    hits = [i for i in df.Hits.tolist() if not np.isnan(i)]
    X = sorted(hits)
    N = len(X)
    Y = np.linspace( 1/N, 1, num=N)
    plt.plot(X,Y)
    plt.xlabel('Hits')
    plt.ylabel('Percentage')
    plt.title(f.split('.')[0].replace("_metadata", ''))

```

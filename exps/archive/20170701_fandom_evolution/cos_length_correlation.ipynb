```python
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import random
from scipy import stats
```

```python
data_path = '../../data/temporal_cos_kudos/prev_futu/'
```

```python
fandom_list_prev = [i for i in os.listdir(data_path) if 'prev' in i]
```

```python
fandom_list_futu = [i for i in os.listdir(data_path) if 'futu' in i]
```

```python
df_1 = pd.read_csv(os.path.join(data_path, fandom_list_prev[0]), sep = '\t').sort()
```

```python
df_1.head()
```

```python
df_2 = pd.read_csv(os.path.join(data_path, fandom_list_futu[0]), sep = '\t').sort()
```

```python
df_2.head()
```

```python
df_1.columns.values
```

```python
df = pd.merge(df_1, df_2, on=['AdditionalTags', 'ArchiveWarnings', 'Author', 'Bookmarks',
       'Category', 'ChapterIndex', 'Chapters', 'Characters', 'Comments',
       'CompleteDate', 'Fandoms', 'Hits', 'Kudos', 'Language', 'Notes',
       'PublishDate', 'Rating', 'Relationship', 'Summary', 'Title', 'URL',
       'UpdateDate', 'Words'], how='inner')
```

```python
len(df_2)
```

```python
len(df)
```

```python
df.head(20)
```

```python
df = df.dropna(subset=['Cos_x', 'Cos_y', 'Kudos'])
x = abs(df['Cos_x'] - df['Cos_y'])
y = df['Kudos']
stats.pearsonr(x,y)
```

```python
for i,fandom in enumerate(fandom_list_futu):
    df = pd.read_csv(os.path.join(data_path, fandom_list_prev[i]), sep = '\t')
    df = df[df.Chapters == 1]
    print(fandom, len(df))
    df = df.dropna(subset=['Cos', 'Words'])
    x = df['Cos']
    y = df['Words']
    print(stats.pearsonr(x,y))
```

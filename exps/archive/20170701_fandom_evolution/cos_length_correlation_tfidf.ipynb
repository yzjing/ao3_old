```python
import pandas as pd
import os
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import random
from scipy import stats
```

```python
data_path = '../../data/tfidf_tocurr_conlen_allave/'
```

```python
fandom_list = [i for i in os.listdir(data_path) if 'tsv' in i]
```

```python
df = pd.read_csv(os.path.join(data_path, fandom_list[0]), sep = '\t')
```

```python
len(df)
```

```python
df.head(20)
```

```python
df = df.dropna(subset=['Cos','Kudos'])
x = df['Cos']
y = df['Kudos']
stats.pearsonr(x,y)
```

```python
for i,fandom in enumerate(fandom_list):
    df = pd.read_csv(os.path.join(data_path, fandom_list[i]), sep = '\t')
    df = df[df.Chapters == 1]
    print(fandom, len(df))
    df = df.dropna(subset=['Cos', 'Words'])
    x = df['Cos']
    y = df['Words']
    print(stats.pearsonr(x,y))
```

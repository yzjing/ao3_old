### tsne clustering

```python
tsne_dict = pickle.load(open(os.path.join(data_path,'shakespare_william_workstemporal_tsne.p'), 'rb'))
```

```python
len(tsne_dict)
```

```python
plt.figure(figsize = (15,5))
for i, time in enumerate(tsne_dict.keys()):
    plt.subplot(2,5,i+1)
    x = list(tsne_dict[time][:,0])
    y = list(tsne_dict[time][:,1])
    for i in range(5):
        x.remove(max(x))
        x.remove(min(x))
    for i in range(5):
        y.remove(max(y))
        y.remove(min(y))
    plt.scatter(x,y)
    plt.xlabel(time.split("_")[0])
plt.tight_layout()
plt.savefig('shakespeare_temporal_tsne.png', type='png')
```
